```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset from sklearn
iris = load_iris()
```

```python
# Create a DataFrame from the dataset
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['target'] = iris.target

# Display the first few rows of the data
print("First few rows of the Iris dataset:")
print(data.head())
```

```python
# Basic Data Manipulation
# Filtering data - for example, selecting rows where 'sepal length (cm)' is greater than 5.0
filtered_data = data[data['sepal length (cm)'] > 5.0]
print("Filtered data (sepal length > 5.0):")
print(filtered_data.head())
```

```python
# Grouping data by the 'target' column and calculating the mean of each group
grouped_data = data.groupby('target').mean()
print("Grouped data (mean of each group):")
print(grouped_data)
```

```python
# Aggregating data - calculating multiple aggregations at once
aggregated_data = data.agg({
    'sepal length (cm)': ['mean', 'sum'],
    'sepal width (cm)': ['min', 'max'],
    'petal length (cm)': ['mean', 'sum'],
    'petal width (cm)': ['min', 'max']
})
print("Aggregated data (mean and sum of sepal and petal dimensions):")
print(aggregated_data)
```

```python
# Analyzing Data
# Descriptive statistics
descriptive_stats = data.describe()
print("Descriptive statistics:")
print(descriptive_stats)
```

```python
# Plotting - Basic plot example
# Plot a histogram of 'sepal length (cm)'
plt.hist(data['sepal length (cm)'], bins=20, alpha=0.7, color='blue')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# Plot a scatter plot between 'sepal length (cm)' and 'petal length (cm)'
plt.scatter(data['sepal length (cm)'], data['petal length (cm)'], alpha=0.7, color='green')
plt.title('Scatter plot between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```
